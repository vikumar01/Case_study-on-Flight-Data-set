```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', 50)
```

```python
df=pd.read_csv("data/arrivals/arrivals.csv",sep='\t')
df
```

```python
df["diff"]= df["flight_arrival_scheduled"] - df["flight_arrival_real"]
```

```python
df['Label']=0
def labelling(raw):
    if(raw >= 0):
        return "On_time"
    elif(raw < 0):
        return "Delayed"
    else:
        return "Cancelled"

df['Label']=df['diff'].apply(labelling)
```

```python
df.head(50)
```

```python
df1=pd.read_csv("data/METAR_airports.csv",sep=',')
df1.head()
```

```python
df["Wind_Speed"]=0
df["Wind_direction"]=0
df["Visibility"]=0
df["Temperature"]=0
df["Dew"]=0
df["Pressure"]=0
```

```python
for i, row in df.iterrows():
    for j, row1 in df1.iterrows():  
        min_time=row1[2]
        max_time=row1[2]+1800
        #print(row[2])
        if (row[13]==row1[0]) & ((min_time <= row[19] <= max_time) or (min_time <= row[17] <= max_time)):
             df.loc[i,'Wind_Speed']=row1['Wind_Speed']
             df.loc[i,'Wind_direction']=row1['Wind_direction']
             df.loc[i,'Visibility']=row1['Visibility']
             df.loc[i,'Temperature']=row1['Temperature']
             df.loc[i,'Dew']=row1['Dew']                               
             df.loc[i,'Pressure']=row1['Pressure']
```

```python
df.to_csv(r'data/arrivals/arrivals_mod.csv',index=False)
```
