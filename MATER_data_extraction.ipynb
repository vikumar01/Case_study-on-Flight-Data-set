```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
pd.set_option('display.max_columns', 50)
from datetime import timezone 
import re
import datetime
```

```python
df=pd.read_csv("data/METAR_all_airports.csv",sep=',')
df
```

```python
df['UTC DATE/TIME'] = pd.to_datetime(df['UTC DATE/TIME'])

for i in range(df.shape[0]):
    dt = df['UTC DATE/TIME'][i]
    utc_time = dt.replace(tzinfo = timezone.utc) 
    df['UTC DATE/TIME'][i] = int(utc_time.timestamp())
  
```

```python
def regex1(raw):
    p=re.search(r"([0-9]{3}|VRB)([0-9]{2,3})G?([0-9]{2,3})?(KT|MPS|KMH)", raw)
    if (p):
        return(p.group())
    else:
        return
df['Wind']=df['METAR'].apply(regex1)
df['Wind'] = df['Wind'].astype(str)
```

```python
def regex1(raw):
    p=re.search(r"([0-9]{3}|VRB)?", raw)
    if (p):
        return(p.group())
    else:
        return
df['Wind_Speed']=df['Wind'].apply(regex1)
```

```python
def regex1(raw):
    p=re.search(r"([0-9]{2}(?=(KT|MPS|KMH)))", raw)
    if (p):
        return(p.group())
    else:
        return
df['Wind_direction']=df['Wind'].apply(regex1)
```

```python
df.drop('Wind', axis = 1, inplace = True)
```

```python
def regex1(raw):
    p=re.search(r"(( [0-9]{4} )|CAVOK|( [0-9./]+\s*SM))", raw)
    if (p):
        return(p.group())
    else:
        return
df['Visibility']=df['METAR'].apply(regex1)
```

```python
def regex1(raw):
    p=re.search(r"(M?[0-9]{2})/M?([0-9]{2})", raw)
    if (p):
        return(p.group())
    else:
        return
df['Temp']=df['METAR'].apply(regex1)
df['Temp'] = df['Temp'].astype(str)
```

```python
def regex1(raw):
    p=re.search(r"(M?[0-9]{2})?", raw)
    if (p):
        return(p.group())
    else:
        return
    
df['Temperature']=df['Temp'].apply(regex1)
```

```python
def regex1(raw):
    p=re.search(r"(?<=/)(M?[0-9]{2})", raw)
    if (p):
        return(p.group())
    else:
        return
    
df['Dew']=df['Temp'].apply(regex1)
```

```python
df.drop('Temp', axis = 1, inplace = True)
```

```python
def regex1(raw):
    p=re.search(r"(A([0-9]{4})|Q([0-9]{4}))", raw)
    if (p):
        return(p.group())
    else:
        return
df['Pressure']=df['METAR'].apply(regex1)
```
